```python
#Create a new column that labels records before 2005 as 'pre-2005' and after 2005 as 'post-2005'
import pandas as pd
```

```python
#set file path to a variable
input_file_path = r"C:\Users\syeda\Documents\vgsales.csv"
```

```python
#read data into our dataframe
Df=pd.read_csv(input_file_path)
Df.head()
```

```python
Df.dtypes
```

```python
Df['Year'] = Df['Year'].apply("int64")
Df['NA_Sales'] = Df['NA_Sales'].apply("int64")
Df['EU_Sales'] = Df['EU_Sales'].apply("int64")
Df['JP_Sales'] = Df['JP_Sales'].apply("int64")
Df['Other_Sales'] = Df['Other_Sales'].apply("int64")
Df['Global_Sales'] = Df['Global_Sales'].apply("int64")
```

```python
Df.dtypes
```

```python
Df.head()
```

```python
Df.loc[(Df["Year"] >= 2005), 'Description'] = "Post 2005"
Df.loc[(Df["Year"] < 2005), 'Description'] = "Pre 2005"
Df
```
